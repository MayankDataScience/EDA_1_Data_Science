````python
# Chart - 4 visualization code
def top_apps_in_category(str):
    str = str.upper()
    top_apps = df1[df1['Category'] == str]
    top_10_apps = top_apps.sort_values(by='Installs', ascending=False).head(10)
    plt.figure(figsize=(15,5), dpi=100)
    str = str.lower()
    plt.title(f'Top 10 Installed Apps Under {str} category',size = 10)  
    graph = sns.barplot(x = top_10_apps.App, y = top_10_apps.Installs, palette= "tab10")
    graph.set_xticklabels(graph.get_xticklabels(), rotation= 45, horizontalalignment='right')
top_apps_in_category('tools')
````
